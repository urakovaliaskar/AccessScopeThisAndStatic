# Wine Classification Using Naive Bayes
The Wine dataset is a well-known dataset in the field of machine learning and statistics. It is frequently used for benchmarking algorithms and learning how to handle data. Here are the key details about the Wine dataset:

## Description:

The Wine dataset consists of 178 samples of wines, each described by thirteen chemical properties. The dataset contains three different classes of wine, each representing a different cultivar from a specific region in Italy.

## Features:

The dataset includes the following thirteen features (all continuous):

1. Alcohol
2. Malic acid
3. Ash
4. Alcalinity of ash
5. Magnesium
6. Total phenols
7. Flavanoids
8. Nonflavanoid phenols
9. Proanthocyanins
10. Color intensity
11. Hue
12. OD280/OD315 of diluted wines
13. Proline

These features represent various chemical properties of the wine samples.

## Target:

The target variable is the class of the wine, which can take one of three possible values:

- Class 0
- Class 1
- Class 2

## Data Structure:

- Number of samples: 178
- Number of features: 13
- Number of classes: 3 (Class 0, Class 1, Class 2)

# Import libraries

Python includes a large number of libraries pre-installed, or available through package managers such as pip, to support machine learning. The libraries we will be using are numpy, pandas and sklearn. Like the previous example we will  be plotting the data using matplotlib, with seaborn used on top of it to better suit our needs.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load the dataset

SKLearn contains the data set with a one-shot load function `load_wine()`. `x` contains the data itself to be analysed, while `y` is the list of targets, the categories for each wine.

```python
# Load the Wine dataset
from sklearn.datasets import load_wine
wine = load_wine()
x = wine.data
y = wine.target
```

# Split the data and standardise it

Just like with the KMC data, we will standardise our data. Before that, however, we will split the data set into two parts, training data and testing data. We split data into training data and test data to evaluate the performance of a machine learning model. The training data is used to fit the model, allowing it to learn patterns and relationships within the data. The test data, which the model has not seen during training, is used to assess how well the model generalizes to new, unseen data. This helps ensure that the model performs well not just on the training data but also on any new data it encounters.

```python
# Split the dataset into training and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

# Standardize the features
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

# Instantiate the model and fit the model to the training data.

In this cell a Gaussian Naive Bayes model (gnb) is created. The model is then trained (fitted) on the training data (x_train, y_train), allowing it to learn the relationship between the features (x_train) and the target labels (y_train).

```python
# Instantiate the Gaussian Naive Bayes model
gnb = GaussianNB()

# Fit the model to the training data
gnb.fit(x_train, y_train)
```

# Make predictions

Now we will use the trained Gaussian Naive Bayes model (gnb) to predict the target labels (y_pred) for the test data (x_test). This allows us to evaluate the model's performance on unseen data.

```python
# Make predictions on the test data
y_pred = gnb.predict(x_test)
```

# Evaluate the model

We can now evaluate the model in three different ways:

- Printing the confusion matrix to show the counts of true positive, true negative, false positive, and false negative predictions.
- Printing the classification report to provide detailed metrics such as precision, recall, and F1-score for each class. Precision is the ratio of correctly predicted positive observations to the total predicted positives. ecall is the ratio of correctly predicted positive observations to all the observations in the actual class. The F1-score is the weighted average of precision and recall.
- Printing the accuracy score to indicate the overall percentage of correct predictions.

```python
# Evaluate the model
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
print("\nAccuracy Score:")
print(accuracy_score(y_test, y_pred))
```

# Plot the confusion matrix

Now we can use seaborn to produce a heatmap visualisation to view the confusion matrix. A confusion matrix is a table used to evaluate the performance of a classification model. It compares the actual target values with the model's predictions. Each row of the matrix represents the instances of the actual class, while each column represents the instances of the predicted class. The main diagonal (top-left to bottom-right) shows the number of correct predictions for each class.

- True Positives (TP): Correctly predicted instances for each class (diagonal elements).
- False Positives (FP): Instances incorrectly predicted as a different class (off-diagonal elements, column-wise).
- False Negatives (FN): Instances incorrectly predicted as not belonging to their actual class (off-diagonal elements, row-wise).

The confusion matrix provides insight into the classification model's performance, highlighting how well the model distinguishes between different classes and identifying any areas where it may be making mistakes.

```python
# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues', xticklabels=wine.target_names, yticklabels=wine.target_names)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
